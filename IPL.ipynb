```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/ipl_matches_smalldata.csv")
```

```python
top_10_players = df.sort_values(by='total_runs', ascending=False).head(10)
```

```python
mean_runs = top_10_players['total_runs'].mean()
```

```python
mode_runs = top_10_players['total_runs'].mode()
```

```python
print("Mean of number of runs scored by top 10 players:")
print(mean_runs)
```

```python
print(f"Mode of runs scored by top 10 players: {mode_runs.values}")
```

```python
filtered_players = df[df['total_runs'] > 500]
print(filtered_players)
```

```python
top_10_players.plot(kind='bar', title='Top 10 Batsmen by Runs', figsize=(10,5))
plt.xlabel('Batsmen') # Corrected the typo from 'xlable' to 'xlabel'
plt.ylabel('Total Runs') # Corrected the typo from 'ylable' to 'ylabel'
plt.show()
```
