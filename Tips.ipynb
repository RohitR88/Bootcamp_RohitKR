```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/tips.csv')
```

```python
smoker_counts = df['Smoking'].value_counts()
```

```python
plt.pie(smoker_counts, labels=smoker_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Percentage of Smokers vs. Non-Smokers')
plt.axis('equal')
```
